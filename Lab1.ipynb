```python
def task_4():
    # Simulating user input: list of integers as a string separated by spaces
    input_string = "5 6 23 8 12 5"  # Simulated input
    
    # Convert the input string into a list of integers
    nums = list(map(int, input_string.split()))
    print(f"Input list: {nums}")
    # Bubble Sort with swap counting
    swap_count = 0
    n = len(nums)
    
    # Bubble sort algorithm
    for i in range(n):
        for j in range(0, n - i - 1):
            if nums[j] > nums[j + 1]:
                # Swap the elements
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                swap_count += 1

    # Output: number of swaps
    print(f"Sorted list: {nums}")
    print(f"There was {swap_count} swaps")

task_4()

```
